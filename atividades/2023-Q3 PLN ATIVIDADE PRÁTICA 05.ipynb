# Imports necessários

```python
import os
import re

# python3 -m pip install requests
from requests import get

# python3 -m pip install langchain
from langchain.chains import LLMChain
from langchain.llms import Petals
from langchain.prompts import PromptTemplate

# python3 -m pip install python-dotenv
from dotenv import load_dotenv 
```

# Localizando o arquivo .env dentro do repo

```python
def find_dotenv() -> str:
    for root, dirs, files in os.walk(str(Path(os.getcwd()).parents[0])):
        if '.env' in files:
            dotenv_path = os.path.join(root, '.env')
            load_dotenv(dotenv_path)
            return dotenv_path
    return None

HUGGINGFACE_API_KEY = os.getenv("HUGGINGFACE_API_KEY")
```

# Função para extração de texto da Wikipedia

```python
def get_wikipedia_content(title : str, print_title = False, print_content = False) -> str:
    url = "https://en.wikipedia.org/w/api.php"
    params = {
        "action": "query",
        "format": "json",
        "titles": title,
        "prop": "extracts",
        "explaintext": True
    }

    response = get(url, params=params)
    
    if response.status_code == 200:
        content = response.json()
        text_content = None
        pages = content['query']['pages']
        for page_id in pages:
            page = pages[page_id]
            if print_title:
                print(f"Title: {page.get('title', 'No title')}")

            text_content = page.get('extract', 'No content').replace('\n', os.linesep)
            if print_content:
                print("Content:")
                print(text_content)
        return text_content
    else:
        return None
```

# Criação da instancia do Petals e initicialização do LLM

```python
llm = Petals(model_name="petals-team/StableBeluga2")
```

### Função para criação de perguntas ao LLM (inglês)

```python
def ask_question(question : str):
    template = """Question: {llm_question} 

    
    Answer: Let's find out: """
    prompt = PromptTemplate(template=template, input_variables=["llm_question"])
    llm_chain = LLMChain(prompt=prompt, llm=llm)
    return llm_chain.run(question)
```

### Teste da conexão ao LLM

```python
print(ask_question("In which year did the first class of students took place in UFABC (Universidade Federal do ABC)?"))
```

## Extração do texto de uma das seções (*Taxation*) da página sobre o Império Romano pela API da Wikipedia

```python
content = get_wikipedia_content("Roman Empire")
print(content)
```

```python
def extract_section(text_content : str, current_topic : str, next_topic : str) -> str:
    pattern = rf"{re.escape(current_topic)}\n(.*?)\n{re.escape(next_topic)}"
    match = re.search(pattern, text_content, re.DOTALL)
    if match:
        return match.group(1).strip()
    else:
        return "pattern not matched"
```

```python
extracted_text = extract_section(content, "=== Taxation ===", "== Economy ==")
print(extracted_text)
```
