#### **INTRODUÇÃO**

**Descrição da API**

A API do GitHub permite que os desenvolvedores interajam com os repositórios, organizações, problemas, pull requests e muitos outros aspectos do GitHub de forma programática. Você pode usar a API do GitHub para automatizar tarefas, integrar o GitHub com outros serviços, criar ferramentas personalizadas e muito mais.

A API do GitHub utiliza o protocolo HTTP e é baseada em REST, o que a torna relativamente fácil de usar com várias linguagens de programação. Ela oferece uma ampla gama de endpoints que você pode acessar para realizar ações como criar, listar, atualizar e excluir repositórios, issues, pull requests e muito mais.

Entende-la possibilita a personalização e a automação de tarefas relacionadas ao desenvolvimento de software, como a integração contínua, a gestão de projetos e a coleta de métricas de código. Isso economiza tempo e esforço, melhora a eficiência e permite que desenvolvedores e equipes personalizem suas interações com o GitHub para atender às necessidades específicas de seus projetos e processos de desenvolvimento.

#### **UTILIZAÇÃO BÁSICA**
A url base é: 'https://api.github.com'

Podemos fazer uma solicitação GET para a url base (endpoint raiz da api) para obter todas as categorias de endpoints com a qual a API REST é compatível

```python
url_base = "https://api.github.com"

response = requests.get(f"{url_base}")
if response.status_code == 200:
    data = response.json()
    formatted_json = json.dumps(data, indent=4)  # Formata o JSON com indentação
    print(formatted_json)
else:
    print(f"Erro na solicitação. Código de status: {response.status_code}")
    print(response.text)
```

Os parâmetros de caminho modificam o caminho da operação. Por exemplo, o caminho "Listar problemas de repositório" é /repos/{owner}/{repo}/issues. As chaves {} denotam parâmetros de caminho que você precisa especificar. Nesse caso, você deve especificar o proprietário e o nome do repositório. Para ver a documentação de referência completa dessa operação, confira "Problemas".

Para obter problemas do repositório octocat/Spoon-Knife, substitua {owner} por octocat e {repo} por Spoon-Knife. Para criar o caminho completo, prepare a URL de base para a API REST do GitHub, https://api.github.com: https://api.github.com/repos/octocat/Spoon-Knife/issues.

```python
response = requests.get(f"{url_base}/repos/octocat/Spoon-Knife/issues")
if response.status_code == 200:
    data = response.json()
    formatted_json = json.dumps(data, indent=4)
    print(formatted_json)
else:
    print(f"Erro na solicitação. Código de status: {response.status_code}")
    print(response.text)

data = {
    "title": "Minha nova issue",
    "body": "Esta é a descrição da issue."
}
```

Abaixo vamos ver informações públicas sobre o usuario "guilhermegbraz" e sobre os repositorios desse usuario

```python
response = requests.get(f"{url_base}/users/guilhermegbraz")
if response.status_code == 200:
    data = response.json()
    formatted_json = json.dumps(data, indent=4)
    print(formatted_json)
else:
    print(f"Erro na solicitação. Código de status: {response.status_code}")
    print(response.text)
```

```python
response = requests.get(f"{url_base}/users/{usuario}/repos")
if response.status_code == 200:
    data = response.json()
    formatted_json = json.dumps(data, indent=4) 
    print(formatted_json)
else:
    print(f"Erro na solicitação. Código de status: {response.status_code}")
    print(response.text)
```

Vimos acima algumas requisisões usando o verbo GET que retornam informações públicas sobre usuarios, repositorios e sobre a própria API do github, entretanto, a maioria dos endpoints necessitam de autenticação para serem utilizados. Sobretudo em requisições de cadastro, modificação ou exclusão de informação (verbos POST, UPDATE e DELETE)


Dito isso, no cabeçalho da requisição podemos passar um token de autenticação, nesse tutorial iremos usar um token de acesso pessoal, e dessa forma seremos capazes de fazer requisições para fazer modificações em repositorios desde que nosso usuario tenha permissões para com esse repositorio especifico.
Por exemplo, vamos listar todos os repositorios, incluindo os privados (que não são listados na requisição vista acima) usando um token pessoal do usuario

```python
headers = {
    'X-GitHub-Api-Version': '2022-11-28',
    "Authorization": f'Bearer {auth_token}'
  }

response = requests.get(f"{url_base}/user/repos", headers=headers)
if response.status_code == 200:
    data = response.json()
    privados = list(filter(lambda x: x["private"] == True, data))
    formatted_json = json.dumps(privados, indent=4)
    print(formatted_json)
else:
    print(f"Erro na solicitação. Código de status: {response.status_code}")
    print(response.text)
```

#### **EXEMPLOS PRÁTICOS**

Vamos ver alguns exemplos práticos usando a API do Github.
A API possui endpoints para qualquer ação que pode ser feita em um repositorio do github, criar issues, pull requests, procurar por repositorios ou usuarios, entre muitas outras.
Nesse exemplo, vamos mostrar operações básicas que podem ser feitas sobre commits de um repositorio e comentarios desse commit. Vamos ver como listar os commits de um repositorio, ler, adicionar, alterar e deletar comentarios nesse commit.

preencha as variaveis `owner` e `repo` com seu usuario e um repositorio, respectivamente, que você pode modificar para executar os exemplos. Os exemplos irão apenas comentar um commit, alterar o comentario e por fim apaga-lo

```python

owner = "abalchiumas"
repo = "pln_2023.3"

headers = {
    'X-GitHub-Api-Version': '2022-11-28',
    "Authorization": f'Bearer {auth_token}'
  }

response = requests.get(f"{url_base}/repos/{owner}/{repo}/commits", headers=headers)
if response.status_code == 200:
    data = response.json()
    formatted_json = json.dumps(data, indent=4)  # Formata o JSON com indentação
    print(formatted_json)
else:
    print(f"Erro na solicitação. Código de status: {response.status_code}")
    print(response.text)
```

Como manipular os comentarios de um commit especifico:

GET

```python
commit_sha = data[0]["sha"]
response = requests.get(f"{url_base}/repos/{owner}/{repo}/commits/{commit_sha}/comments", headers=headers)
data = response.json()
formatted_json = json.dumps(data, indent=4)
if response.status_code == 200:
    print(formatted_json)
else:
    print(f"Erro na solicitação. Código de status: {response.status_code}")
    print(formatted_json)
```

POST

```python
body = {
    "body":"Excelente começo para a atividade 3"
}
response = requests.post(f"{url_base}/repos/{owner}/{repo}/commits/{commit_sha}/comments", headers=headers, json=body)
data = response.json()
formatted_json = json.dumps(data, indent=4)
if response.status_code == 201:
    print(formatted_json)
else:
    print(f"Erro na solicitação. Código de status: {response.status_code}")
    print(formatted_json)

# Executando novamente a celula anterior, agora a contagem de comentarios no commit deve ter se alterado
```

UPDATE

Com o comentario criado anteriormente, vamos altera-lo e depois exclui-lo

```python
id_comentario_cirado = data["id"]
body = {
    "body":"Excelente começo para a atividade 3. Mas vamos modificar esse comentário"
}

response = requests.patch(f"{url_base}/repos/{owner}/{repo}/comments/{id_comentario_cirado}", headers=headers, json=body)
data = response.json()
formatted_json = json.dumps(data, indent=4)
if response.status_code == 200:
    print(formatted_json)
else:
    print(f"Erro na solicitação. Código de status: {response.status_code}")
    print(formatted_json)
```

DELETE

```python

response = requests.delete(f"{url_base}/repos/{owner}/{repo}/comments/{id_comentario_cirado}", headers=headers, json=body)
if response.status_code == 204:
    print(f"Comentario de id: {id_comentario_cirado} excluido")
else:
    print(f"Erro na solicitação. Código de status: {response.status_code}")
    print(response.text)
```
